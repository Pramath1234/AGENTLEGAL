```python
from langchain_huggingface import HuggingFaceEmbeddings, HuggingFaceEndpoint
from langchain.vectorstores import Pinecone as LangchainPinecone
from langchain.chains import RetrievalQA
import os
from pinecone import Pinecone, ServerlessSpec
import time
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge import Rouge
from bert_score import score

# Download the necessary nltk package
nltk.download('punkt')
# Setup API tokens
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "HUGGINGFACEHUB_API_TOKEN"
os.environ["PINECONE_API_KEY"] = "PINECONE_API_KEY"
os.environ["PINECONE_ENV"] = "PINECONE_ENV"

# Initialize Pinecone using the Pinecone class
pc = Pinecone(api_key=os.environ.get("PINECONE_API_KEY"))

# Check if the index exists, create it if not
index_name = "atmecs"
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=768,  # Adjust to your embedding model's dimension
        metric='cosine',  # Choose your preferred metric (cosine, euclidean, etc.)
        spec=ServerlessSpec(
            cloud='aws',
            region=os.environ["PINECONE_ENV"]
        )
    )

# Initialize the embedding model
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Initialize Langchain Pinecone Vector Store
vectorstore = LangchainPinecone.from_existing_index(index_name=index_name, embedding=embedding_model)

# Initialize the LLM
llm = HuggingFaceEndpoint(repo_id="mistralai/Mistral-Small-Instruct-2409")

# Initialize the RetrievalQA chain with embeddings and LLM
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever())

# Accept user query
user_query = input("Please enter your question: ")
# Step 1: Embed the user query using the embedding model
query_embedding = embedding_model.embed_query(user_query)
instruction = "You are a highly knowledgeable legal assistant. You have a good knowledge of indian legal system and corporate law. Your primary role is to assist users by answering their questions related to legal matters, contracts, compliance, and court cases.Provide accurate and concise answers based on your understanding of legal concepts and practices.If a question is ambiguous or outside your expertise, respond with I don't know instead of attempting to guess.Always maintain a professional tone, and structure your answers clearly.Feel free to ask for clarification if a users question is unclear."

query_with_instruction = instruction +" "+ user_query
start_time = time.time()
response = qa_chain.run(query=query_with_instruction, max_length=500, return_only_outputs=True)
end_time = time.time()
response_time = end_time - start_time
# Step 4: Print the helpful answer
print(response)
print("Response Time (seconds):", response_time)
references = [
    nltk.word_tokenize("A limited liability company (LLC) is a business entity that protects the personal assets of its owners."),
    nltk.word_tokenize("An LLC shields its members from being personally responsible for the company's debts or liabilities."),
    nltk.word_tokenize("The owners of a limited liability company are not personally liable for the financial obligations of the business."),
    nltk.word_tokenize("An LLC is a hybrid entity combining features of both a corporation and a partnership."),
    nltk.word_tokenize("Forming an LLC involves filing the articles of incorporation and creating an operating agreement."),
    nltk.word_tokenize("Members of an LLC enjoy limited liability protection, separating personal assets from business debts."),
    nltk.word_tokenize("LLCs can choose to be taxed as either a partnership or a corporation depending on the preference of the members."),
    nltk.word_tokenize("A limited liability company (LLC) protects its owners from personal financial risk beyond their investment in the company."),
    nltk.word_tokenize("The Limited Liability Partnership Act, 2008 (LLP Act) is a legislation in India that provides for the formation and regulation of Limited Liability Partnerships (LLPs)."),
    nltk.word_tokenize("An LLP is a separate legal entity, formed and incorporated under the LLP Act, with a legal existence distinct from its partners."),
    nltk.word_tokenize("The liability of the partners is limited to their agreed contribution to the LLP, similar to shareholders in a company. This means that partners are not personally liable for the LLP's debts or liabilities."),
    nltk.word_tokenize("An LLP is formed by filing the incorporation document with the Registrar of Companies (ROC) and obtaining a Certificate of Incorporation."),
    nltk.word_tokenize("Corporate bylaws are the internal rules that govern the operations of a corporation. They define the roles of the board of directors, how meetings are conducted, the rights and responsibilities of shareholders, and procedures for corporate actions like mergers. Bylaws ensure compliance with state law and provide a framework for corporate decision-making."),
    nltk.word_tokenize("Incorporating a company involves filing articles of incorporation with the state's Secretary of State office, choosing a business name, selecting a board of directors, and defining the company’s share structure."),
    nltk.word_tokenize("An LLC (Limited Liability Company) offers flexibility in taxation and management but isn't technically a corporation. An S-Corp allows income to pass through to shareholders, avoiding double taxation, but has limitations on the number and type of shareholders. A C-Corp is taxed separately from its owners and can have an unlimited number of shareholders but faces double taxation—on both corporate profits and dividends."),
    nltk.word_tokenize("Articles of incorporation are legal documents filed with a government body to legally document the creation of a corporation."),
    nltk.word_tokenize("Corporate governance refers to the system by which companies are directed and controlled. It involves a set of rules, practices, and processes ensuring accountability, fairness, and transparency in a company's relationship with stakeholders. Good governance prevents conflicts of interest and ensures the integrity of business operations."),
    nltk.word_tokenize("CSR laws encourage companies to operate ethically and consider their environmental and social impacts. In India, for instance, companies meeting certain financial thresholds are required to spend at least 2% of their average net profit on CSR activities. CSR initiatives include environmental conservation, education, and poverty alleviation efforts."),
    nltk.word_tokenize("Patentable subject matter includes new and useful processes, machines, articles of manufacture, or compositions of matter. In some jurisdictions, certain subjects—such as abstract ideas, natural phenomena, and mathematical formulas—are not patentable. The invention must also be novel, non-obvious, and have a useful application."),
    nltk.word_tokenize("Trade secrets are protected through confidentiality agreements, restricted access to sensitive information, and internal security protocols. If someone improperly discloses or uses a trade secret, the owner can pursue legal remedies such as injunctions or damages under trade secret laws."),
    nltk.word_tokenize("Copyright law protects original works of authorship, such as software, written documents, and designs. In corporate environments, businesses must ensure that employees or contractors assign copyright ownership of work created for the company. Unauthorized use of copyrighted material can lead to infringement claims."),
    nltk.word_tokenize("Licensing agreements grant permission to a licensee to use intellectual property (IP), such as patents or trademarks, owned by the licensor. The agreement outlines terms such as the scope of use, royalties, exclusivity, and duration. It helps protect the licensor's IP while allowing the licensee to commercialize it."),
    nltk.word_tokenize("A compliance officer ensures that a company adheres to legal standards and internal policies. They monitor regulatory developments, conduct audits, and create compliance programs. Their role is crucial in preventing legal violations, reducing risks, and ensuring the company's operations align with applicable laws."),
    nltk.word_tokenize("Securities laws regulate the issuance and sale of company shares to protect investors. Before issuing stock, companies must file disclosures with the SEC or relevant authorities, providing financial information and material facts about the business. These laws ensure transparency and prevent fraud in capital markets."),
    nltk.word_tokenize("Corporate tax regulations vary by jurisdiction but typically involve taxes on profits, capital gains, and payroll. Common tax planning strategies include using tax deductions, credits, and deferrals. Companies may exploit legal loopholes, such as transfer pricing or establishing subsidiaries in low-tax jurisdictions, to minimize tax liabilities."),
    nltk.word_tokenize("Employers are required to adhere to labor laws that govern minimum wage, overtime, employee safety, and non-discrimination practices. These laws ensure fair treatment of employees, compliance with health and safety standards, and the prevention of workplace harassment. Employers are also responsible for paying mandatory benefits, such as social security or pensions."),
    nltk.word_tokenize("Employee termination requires adherence to legal processes to avoid wrongful termination claims. Employers must provide sufficient notice, document reasons for termination, and offer severance if required. Discriminatory or retaliatory terminations are illegal and can lead to lawsuits and penalties"),
    nltk.word_tokenize("Workplace harassment laws protect employees from unwanted conduct based on characteristics such as race, gender, religion, or sexual orientation. These laws prohibit verbal, physical, or visual harassment and create obligations for employers to maintain a harassment-free environment through training, policies, and complaint procedures."),
    nltk.word_tokenize("Wage and hour laws regulate employee compensation, including minimum wage, overtime pay, and working hours. These laws vary by jurisdiction but typically require employers to pay at least a minimum wage and overtime for hours worked beyond a certain threshold (e.g., 40 hours per week)."),
    nltk.word_tokenize("Labor unions represent employees in negotiations with employers on issues like wages, working conditions, and benefits. Collective bargaining agreements (CBAs) formalize the terms agreed upon by both parties. Employers must follow the terms of the CBA and work collaboratively with union representatives to resolve disputes."),
    nltk.word_tokenize("Corporations accused of criminal conduct may defend against charges by proving lack of intent, challenging the evidence, or demonstrating compliance with the law. For instance, if a company can show it had a robust compliance program in place to prevent fraud, it might mitigate criminal liability or negotiate for reduced penalties."),
    nltk.word_tokenize("A Limited Liability Company (LLC) is a business structure that combines the limited liability protection of a corporation with the tax benefits and flexibility of a partnership. In an LLC, the owners, known as members, are not personally liable for the company's debts or legal obligations. LLCs are popular for small businesses because they offer flexibility in management and tax treatment, with fewer compliance requirements than corporations."),
    nltk.word_tokenize("To form an LLC, the first step is to choose a business name that complies with your state's LLC naming rules. Then, you must file the Articles of Organization with the state, typically with the Secretary of State’s office."),
    nltk.word_tokenize("The primary difference between an LLC and a corporation is in ownership structure and taxation."),
    nltk.word_tokenize("Forming an LLC offers several benefits. One key advantage is limited liability protection, which means that members' personal assets are shielded from business debts or legal judgments. Additionally, LLCs offer flexibility in taxation, allowing members to choose how they want to be taxed—either as a pass-through entity or as a corporation. "),
    nltk.word_tokenize("An Operating Agreement is an internal document that outlines the ownership, management, and operational procedures of an LLC. While not always required by law, it is highly recommended for multi-member LLCs."),
    nltk.word_tokenize("LLCs can choose between two types of management structures: member-managed or manager-managed. "),
    nltk.word_tokenize("If an LLC member wants to leave the company, the process is typically governed by the terms outlined in the Operating Agreement. The agreement may include buyout provisions, detailing how the member’s ownership interest will be valued and transferred. In the absence of an Operating Agreement, state default laws will apply, which may require the LLC to dissolve if members cannot agree on how to proceed. It is essential to have clear provisions for member withdrawal to avoid disputes and ensure business continuity."),
    nltk.word_tokenize("A single-member LLC has only one owner, while a multi-member LLC has two or more owners, known as members. Both types of LLCs offer limited liability protection and flexibility in management. The primary difference lies in taxation: a single-member LLC is treated as a disregarded entity for tax purposes, meaning it is taxed like a sole proprietorship, with income and losses reported on the owner's personal tax return. "),
    nltk.word_tokenize("In India, a Limited Liability Company (LLC) can operate under two primary types of management structures:1. **Member-Managed LLC**: In this structure, the members (owners) of the LLC manage the day-to-day operations and decision-making processes. Each member has equal rights and responsibilities in managing the company. This structure is typically used when the LLC has a small number of members.2. **Manager-Managed LLC**: In this structure, the members appoint one or more managers (who can be members or non-members) to handle the LLC's operations and make decisions on behalf of the company. This structure is common in larger LLCs where it's not feasible for all members to be involved in management. The managers have the authority to bind the LLC, but they are still accountable to the members.")
]

# Step 3: Tokenize the generated response
candidate = nltk.word_tokenize(response)

# Step 4: Calculate the BLEU score
smooth_fn = SmoothingFunction().method1  # Smoothing for better score
bleu_score = sentence_bleu(references, candidate, smoothing_function=smooth_fn)
rouge = Rouge()
# Convert references and candidate to strings for ROUGE
references_joined = [" ".join(ref) for ref in references]
candidate_joined = " ".join(candidate)

# Compute ROUGE for each reference and average
rouge_scores = [rouge.get_scores(candidate_joined, ref)[0] for ref in references_joined]
avg_rouge = {
    'rouge-1': {
        'precision': sum([score['rouge-1']['p'] for score in rouge_scores]) / len(rouge_scores),
        'recall': sum([score['rouge-1']['r'] for score in rouge_scores]) / len(rouge_scores),
        'f1': sum([score['rouge-1']['f'] for score in rouge_scores]) / len(rouge_scores),
    },
    'rouge-2': {
        'precision': sum([score['rouge-2']['p'] for score in rouge_scores]) / len(rouge_scores),
        'recall': sum([score['rouge-2']['r'] for score in rouge_scores]) / len(rouge_scores),
        'f1': sum([score['rouge-2']['f'] for score in rouge_scores]) / len(rouge_scores),
    },
    'rouge-l': {
        'precision': sum([score['rouge-l']['p'] for score in rouge_scores]) / len(rouge_scores),
        'recall': sum([score['rouge-l']['r'] for score in rouge_scores]) / len(rouge_scores),
        'f1': sum([score['rouge-l']['f'] for score in rouge_scores]) / len(rouge_scores),
    }
}


# Duplicate the candidate response for each reference to make sure they are equal in length
candidate_responses = [response] * len(references_joined)

# Compute BERTScore
P, R, F1 = score(candidate_responses, references_joined, lang="en", rescale_with_baseline=True)
# Print all scores
print(f"BLEU Score: {bleu_score}")
print(f"ROUGE Scores: {avg_rouge}")
print(f"BERTScore - Precision: {P.mean().item()}, Recall: {R.mean().item()}, F1: {F1.mean().item()}")
```
